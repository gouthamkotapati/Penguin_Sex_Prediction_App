```python
df["culmen_depth_mm"]=df["culmen_depth_mm"].fillna(df["culmen_depth_mm"].mean())
```

```python
df["flipper_length_mm"]=df["flipper_length_mm"].fillna(df["flipper_length_mm"].mean())
```

```python
df["body_mass_g"]=df["body_mass_g"].fillna(df["body_mass_g"].median())
```

```python
df["sex"]=df["sex"].replace(".","MALE")
```

```python
df["sex"]=df['sex'].fillna("MALE")
```

```python
df["species"]=df["species"].replace({"Adelie":0,"Gentoo":1,"Chinstrap":2})
```

```python
df["island"]=df["island"].replace({"Biscoe":0,"Dream":1,"Torgersen":2})
```

```python
df["sex"]=df["sex"].replace({"MALE":0,"FEMALE":1})
```

```python
df
```

```python
x=df.drop("sex",axis=1)
y=df['sex']
```

```python
x[0:10]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
y_train
```

```python
y_test[0:10]
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
y_pred=tree.predict(x_test)
```

```python
y_pred[1:10]
```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score
test_acc=accuracy_score(y_test,y_pred)
```

```python
print("accuracy_score",test_acc)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```
